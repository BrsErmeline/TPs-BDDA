**Question 1** : Ecrire une procédure qui permet de prendre en paramètre une liste de dépendances fonctionnelles et les affiche.

```python
def print_dependencies(dependencies):
  for dependency in dependencies:
    left, right = dependency
    left_item = ','.join(sorted(left))
    right_item = ','.join(sorted(right))
    print(f'{left_item} -> {right_item}')
```

```python
mydependencies = [
[ {'A'}, {'B'} ],
[ {'A'}, {'C'} ],
[ {'C', 'G'}, {'H'} ],
[ {'C', 'G'}, {'I'} ],
[ {'B'}, {'H'} ]
]
```

```python
print_dependencies(mydependencies)
```

**Question 3** :  Ecrire une fonction qui, prenant en paramètre un ensemble inputset, revoie tous ces sous ensembles. Rappelons qu’un sous-ensemble S d’un ensemble E est un ensemble tel que S ⊆ E.
L’ensemble qui a pour éléments tous les sous-ensemble de E est noté P(E). Et card(P(E)) = 2n, où n = card(E)(la cardinalité de E). N’oubliez pas l’instruction permettant d’importer itertools.

```python
import itertools
```

```python
def power_set(inputset):
  sous_ens_list = []
  for i in range(1, len(inputset) + 1):
    for subset in itertools.combinations(inputset, i):
      sous_ens_list.append(set(subset))
  return sous_ens_list
```

```python
exemple = {'A', 'B', 'C'}
```

```python
print("Power set : ", power_set(exemple))
```

Question 4 :  Ecrire une fonction qui permet, étant donné un ensemble de dépendances fonctionnelles F et
un ensemble d’attributs K, de retourner la fermeture (clôture) de K.

```python
def closure(attributes, dependencies):
  closure_set = set(attributes)
  changed = True
  while changed:
    changed = False
    for left, right in dependencies:
      if left.issubset(closure_set) and not right.issubset(closure_set):
        closure_set.update(right)
        changed = True
  return closure_set
```

```python
print(attributes, " + = ", closure(attributes, mydependencies))
```

```python
def print_closureF(dependencies, relations):
    resultat = []
    for subset in power_set(relations):
        subset_closure = closure(subset, dependencies)
        for attr in subset_closure:
            if attr not in subset:
                resultat.append([subset, {attr}])

    return resultat
```

**Question 6** : Ecrire une fonction qui permet, étant donnée un ensemble de dépendances fonctionnelles F et deux ensembles d’attributs α et β , de retourner vrai si α détermine fonctionnement β.

```python
def determinates_functionally(dependencies, alpha, beta):
  closure_alpha = closure(alpha, dependencies)
  return beta.issubset(closure_alpha)
```

**Question 8** : Ecrire une fonction qui permet, étant donnée un ensemble de dépendances fonctionnelles F, une relation R et un ensemble d’attributs K, de retourner vrai si K est une clé candidate.

```python
def isCandidateKey(dependencies, relation, K):
  if not est_supercle(dependencies, relation, K):
    return False
  if not est_supercle(dependencies , relation , K ): return False
  for A in K:
    K_aux = set(K)
    K_aux.discard(A)
    if est_supercle(dependencies, relation, K_aux):
      return False
  return True
```

```python
def calculate_all_subsets(relations):
  subsets = [set()]
  for element in relations:
    new_subsets = []
    for subset in subsets:
      new_subsets.append(subset.union({element}))
    subsets.extend(new_subsets)
  return subsets
```

**Question 12** : Ecrire une fonction qui permet, étant donnée une relation R et un ensemble de dépendances fonctionnelles F, de retourner vrai si cette relation est en BCNF.

```python
def isBCNF(dependencies, relations):
    for alpha, beta in dependencies:
        if alpha.issubset(relations) and beta.issubset(relations):
            if not est_supercle(dependencies, relations, alpha):
                return False
    return True
```

**Question 13** :  Ecrire une fonction qui permet, étant donnée un ensemble de relations T et une liste de dépendances fonctionnelles F, de retourner vrai si le schéma défini par ces relations est en
BCNF.

```python
def schemaBCNF(dependencies, relations):
    for R in relations:
        if not isBCNF(dependencies, R):
            return False
    return True
```

**Question 14** : Ecrire une fonction qui permet, étant donnée un ensemble de dépendances fonctionnelles F et un ensemble de relations T, d’implémenter l’algorithme de décomposition en BCNF, vu en
cours.

```python
def decomposeBCNF(dependencies, relations):
    resultat = relations.copy()
    changed = True
    while changed:
        changed = False
        for R in resultat:
            for alpha, beta in dependencies:
                if alpha.issubset(R) and beta.issubset(R):
                    if not est_supercle(dependencies, R, alpha):
                        R1 = alpha | beta
                        R2 = R - (beta - alpha)
                        resultat.remove(R)
                        resultat.append(R1)
                        resultat.append(R2)
                        changed = True
                        break
            if changed:
                break
    return resultat
```
